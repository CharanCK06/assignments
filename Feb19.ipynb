```python
from flask import Flask
```

```python
#1
'''
- Flask is python library or web application framework using which we can create
API's

Advantages:
Lightweight
Flexible
compatible
'''
```

```python
#2
app = Flask(__name__)
@app.route("/hello")
def hello():
    return "<h1>“HEllo World!!!!”</h1>"
if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#3
'''
 app routing refers to the process of mapping URLs to view functions.
 
 We use app routes in Flask to define the URLs that our web application will respond to, and to specify the view functions that will handle those requests.
'''
```

```python
#4
@app.route("/welcome")
def welcome():
    return "<h1>“Welcome to ABC Corporation”</h1>"

@app.route("/")
def display():
    return "Company Name: ABC Corporation\nLocation: India\nContact Detail: 999-999-9999"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#5

from flask import url_for

app = Flask(__name__)

@app.route('/hello1/<name>')
def hello1(name):
    return f'Hello, {name}!'

@app.route('/u')
def index1():
    # URL for the hello endpoint with name argument set to 'John'
    url = url_for('hello1', name='John')
    return f'The URL for hello John is: {url}'

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```
