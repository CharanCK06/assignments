```python
#1
'''
Label Encoding: Label Encoding assigns a unique numerical label to each category in the variable.
                It simply replaces the categories with integers starting from 0 to (number of categories - 1).
                
                For example:
                Categories: "Red", "Blue", "Green"
                Label Encoding: 0, 1, 2
                
                The order of the labels is arbitrary and does not imply any inherent order or magnitude among 
                the categories.

Ordinal Encoding: Ordinal Encoding assigns numerical labels to the categories based on their order or rank.
                  It considers the ordinal relationship or relative order of the categories.
                 
                  For example:
                  Categories: "Low", "Medium", "High"
                  Ordinal Encoding: 0, 1, 2
                  
                  Here, the labels are assigned based on the ordinal relationship of the categories, indicating 
                  that "High" has a higher value than "Medium," and "Medium" has a higher value than "Low."
```

```python
#2
'''
    Target Guided encoding is a type of encoding that encodes the categorical value based on the target value.
    - It replaces the categorical value with mean or median of the target variable.
'''
import seaborn as sns
df = sns.load_dataset('tips')
df
```

```python
mean_time = df.groupby('time')['total_bill'].mean().to_dict()
mean_time
```

```python
df['time_mean'] = df['time'].map(mean_time)
df
```

```python
df = pd.DataFrame({'Color' : ['red', 'green', 'blue'],
                   'Size' : ['small', 'medium','large'],
                   'Material' : ['wood', 'metal', 'plastic']})
df
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['Color']]).toarray()
encoder_df = pd.DataFrame(encoded,columns = encoder.get_feature_names_out())
encoder_df
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories = [['small', 'medium',
'large']])
encoder.fit_transform(df[['Size']])
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit_transform(df['Material'])
```

```python
#7
weather = pd.DataFrame({'temperature' : [25, 30, 35, 40, 45],
                        'humidity' : [60, 65, 70, 75, 80],
                        'weather_condition' : ['Sunny', 'Cloudy', 'Rainy','Sunny','Cloudy'],
                        'wind_direction' : ['North', 'South','East', 'West', 'North']
                       })
weather
```

```python
weather.cov()
```

```python
encode = LabelEncoder()
encoded = encoder.fit_transform(weather[['weather_condition']])
weather['weather_con'] = encoded
encoded = encoder.fit_transform(weather[['wind_direction']])
weather['weather_dir'] = encoded
weather
```

```python
weather.cov()
```
