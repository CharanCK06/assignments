```python
#1
df.iloc[1]
```

```python
#2
'''
loc - locates data from dataframe according to user-defined indexing.
iloc - locates data from dataframe according to system-generated indexing.
'''
```

```python
#3
reindex = [3,0,1,2]
new_df = df.reindex(reindex)
new_df
```

```python
new_df.loc[2]
```

```python
new_df.iloc[2]
```
