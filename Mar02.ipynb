```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1
'''
matplotlib is a python library used to analyse and visualize the data.
5plots:
-line plots
-bar plots
-scatter plots
-histogram
-piechart
'''
```

```python
#2
'''
Scatter plot is used to identify trends and clusters in data
'''
```

```python
#2
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#3
'''Subplots are used to plots multiple graphs in single figure'''
```

```python
#3
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(4,1,1)
plt.plot(x,y)
plt.title('1')

y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(4,1,2)
plt.plot(x,y)
plt.title('2')

y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(4,1,3)
plt.plot(x,y)
plt.title('3')

y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(4,1,4)
plt.plot(x,y)
plt.title('4')

plt.show()
```

```python
#4
'''
Bar plots use rectangular bars to represent the data and are useful for comparing values between different groups
'''
```

```python
#4
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
```

```python
#5
'''
 Box plot display the summary of the set of data values having 
 properties like minimum, first quartile, median, third quartile and maximum
'''
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box1)
```
