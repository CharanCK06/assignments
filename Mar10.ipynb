```python
#1
'''
Estimation statistics is a branch of statistics that deals with the process of estimating an unknown parameter of a population
based on a sample of data.

-Point estimate: It is a single value that is used to estimate the value of an unknown parameter. 
                 It does not provide information about the accuracy of the estimate or the level of uncertainty associated with it.
                 
-Interval estimate: It provides a range of values that is likely to contain the true value of the population parameter. 
                    This range is called a confidence interval and it is calculated based on the sample data and a specified level of confidence.
'''
```

```python
#2
import math
def estimate_pmean(x,std,n):
    z_score = 1.96
    moe = z_score * (std / math.sqrt(n))
    lower = x - moe
    upper = x + moe
    return(lower,upper)
x = float(input("Enter the Sample Mean: "))
std = float(input("Enter the Standard Deviation:"))
n = int(input("Enter the Sample Size: "))
print(estimate_pmean(x,std,n))
```

```python
#3
'''
Hypothesis testing is a statistical method used to make decisions about a population based on sample data.

Hypothesis testing is used to answer research questions, test theories, and make decisions based on data.

The importance of hypothesis testing lies in its ability to provide a framework for making decisions based on data. 
By using statistical techniques to evaluate the evidence, we can make more informed decisions, avoid bias and error, 
and minimize the risk of drawing incorrect conclusions. 
'''
```

```python
#4
'''
Null Hypothesis(H0): There is no significant difference between the average weight of male college students and female college students.

Alternative Hypothesis(H1): The average weight of male college students is greater than the average weight of female college students.
'''
```

```python
#5
import numpy as np
from scipy.stats import ttest_ind
sample1 = np.random.normal(loc=10, scale=2, size=50)
sample2 = np.random.normal(loc=12, scale=2, size=50)
test_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)
if p_value < 0.05:
    print("Reject the null hypothesis - there is evidence of a significant difference.")
else:
    print("Fail to reject the null hypothesis - there is insufficient evidence of a significant difference.")
```

```python
# 9
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
df = 10  # degrees of freedom
x = np.linspace(-4, 4, 1000)  # generate x values
y = stats.t.pdf(x, df)  # calculate the y values using the t probability density function

plt.plot(x, y)  # plot the t-distribution
plt.title("Student's t-distribution (df=10)")
plt.xlabel("x")
plt.ylabel("Probability density")
plt.show()
```

```python
#12
'''

The t-statistic is a test statistic that measures the difference between the sample mean and the population mean, 
relative to the standard error of the mean.

t = (x̄ - μ) / (s / √n)

where:

x̄ is the sample mean
μ is the population mean (hypothesized value)
s is the sample standard deviation
n is the sample size
'''
```
